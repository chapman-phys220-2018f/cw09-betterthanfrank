## Graphing different derivatives

**Discussing the problem**

In this problem we are taking the derivative of some function f(x). The derivative of this function f at some point x is defined as the limit:

$$f'(x) = \frac{df}{dx}(x) = Df(x) \equiv \lim_{h \to 0} \frac{f(x+h) - f(x)}{(x + h) - x} = \lim_{h\to 0} \frac{f(x) - f(x - h)}{x - (x - h)} = \lim_{h \to 0} \frac{f(x+h) - f(x-h)}{(x+h)-(x-h)}$$

These are all used approximating the derivative. The central difference equation is the most accurate for approximating the derivative, but we need to use the forward and backward difference equations as well for coding the gradient in this problem.
